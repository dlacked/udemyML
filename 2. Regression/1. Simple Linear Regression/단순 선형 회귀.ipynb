# 데이터 전처리

```python
import numpy as np
import pandas as pd

df = pd.read_csv('Salary_Data.csv')
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

# 단순 선형 회귀

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression() #단순 선형 회귀 객체
regressor.fit(X_train, y_train) #단순 선형 회귀 학습

y_pred = regressor.predict(X_test) #테스트 독립 변수를 가지고 종속 변수 예측
```

```python
import matplotlib.pyplot as plt

plt.scatter(X_test, y_test, color='red') #산점도 지정
plt.plot(X_test, regressor.predict(X_test), color='blue') # 선형 그래프 지정
plt.title('Salary vs Experience (Test Set)')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.show()
```
