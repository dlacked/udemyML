# K-Means Clustering

## Data Preprocessing

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('Mall_Customers.csv')

X = df.iloc[:, [3, 4]].values
# CustomerID는 신원 정보일 뿐이므로 제외해야 함
# 패턴을 파악하기 위해 남겨야 할 열은 Annual Income과 Spending Score 뿐임
# 종속 변수는 있을 수가 없어서 Y는 제외해야 함
```

## Using the elbow method to find the optimal number of clusters

```python
from sklearn.cluster import KMeans

wcss = []

for i in range(1, 11): 
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 0)
    # 군집 10개 파악 (WCSS)
    # k-means++ 무작위 초기화 방지
    kmeans.fit(X)
    wcss.append(kmeans.inertia_) # WCSS 값을 구할 수 있음

plt.plot(range(1, 11), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show # 군집 최적 개수 5개
```

## Training the K-Means model on the dataset

```python
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = 0)
y_kmeans = kmeans.fit_predict(X) # 종속 변수 반환
print(y_kmeans) # 각 고객에 해당하는 서로 다른 군집을 확인할 수 있음
```

## Visualising the clusters

```python
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], color='red', label='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], color='orange', label='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], color='green', label='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], color='blue', label='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], color='purple', label='Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], color='black', s = 150, label='centroid')
plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
