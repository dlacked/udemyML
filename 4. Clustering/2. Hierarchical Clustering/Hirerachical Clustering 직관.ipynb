# What HC does for you

HC 또한 K-means와 비슷한 결과를 가지나 다른 과정을 거쳐 결과를 생성한다.

HC는 두 타입의 과정이 존재한다.

Agglomerative(응집적) 또는 Divisive(분할적)이다.

응집적인 경우는 아래부터 올라오는 방식이고, 분할적인 경우는 위에서 내려오는 방식이다.

# Agglomerative HC

1. 각 데이터 포인트를 단독 포인트 군집으로 만들어서 $$n$$개의 군집 만드는 것이다.

즉, 처음엔 모든 각각의 데이터 포인트들이 서로 독립적인 군집을 가질 것이다.

2. 가장 가까운 두 데이터 포인트를 가져와서 그걸 한 군집으로 만들어 두 개를 하나의 군집으로 만든다. -> $$n-1$$개의 군집이 형성됨

3. 가장 가까운 두 군집들을 지금 있는 것들 중 골라서 그걸 하나의 군집으로 만들어 $$n-2$$의 군집을 만든다.

4. 3단계를 반복해서 **한 개의 군집만 남을 때까지** 반복한다.

## Distance Between Clsuters

3단계를 행하기 위해선 두 군집 사이의 거리를 측정해야 한다.

군집 사이의 거리를 측정하는 방법은 다양하게 있을 수 있는데, 이를 뭘로 정의하는지가 HC에 있어 매우 중요하다.

하지만 결국 모든 단계를 마치면 한 개의 군집만 남아버리는데, HC의 목적이 정확히 뭘까?

## How Do Dendrograms Work?

Dendrogram은 일종의 HC 알고리즘 메모리 역할을 한다.

우리가 수행하는 단계들을 모두 Dendrogram으로 기록하여, 군집이 거리가 얼마나 먼 군집이랑 합쳐졌는지를 알 수 있다.

이제 이를 어떻게 써서 계층적 클러스터링 알고리즘을 향상하는 방법을 알아 볼 것이다.


## Dendrograms

Dendrogram의 가로 수준을 보고 비유사성 한계점을 설정하는 것이다.

세로축은 포인트들 사이의 중앙값 거리를 측정해서 포인트들이나 클러스터들 사이 비유사성을 상징한다.

모든 비유사성에 대한 한계점을 설정하면 클러스터 중 한계점을 넘지 않는 수준에서 몇 개의 클러스터가 나오는지 확인하면 된다.  

그저 가로(비유사성) 한계선이 얼마나 많은 세로선과 교차하는지만 확인하면 된다.

그러면 우리가 비유사성 한계점을 어떻게 설정해야 최적한 군집이 나올까?

가장 표준적인 접근 중 하나는 바로 가장 높은 세로 거리를 찾는 것이다.

높은 세로 거리를 가로지르는 비유사성 한계선을 그으면 가장 최적의 군집이 나올 수 있다.
