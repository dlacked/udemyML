# Hierarchical Clustering

## Data Preprocessing

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('Mall_Customers.csv')

X = df.iloc[:, [3, 4]].values
```

## Using the dendrogram to find the optimal number of clusters

```python
import scipy.cluster.hierarchy as sch
# 덴드로그램을 만들기 위해 호출하는 함수를 가진 모듈

dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))
# linkage의 첫 번째 인자: 클러스터를 파악하고자 하는 Feature Matrix
# linkage의 두 번째 인자: 클러스터를 얻는 방법

plt.title('Dendrogram')
plt.xlabel('Customers')
plt.ylabel('Euclidean Distance')
plt.show()
```

가장 관련성 높은 클러스터를 얻는 방법은 최소 평균편차를 이용하여 계산하는 것이다.

계층적 군집화에서 얻은 각 클러스터에서의 편차를 최소한으로 만들어준다.

최소 평균편차를 이용하기 위해 `method` 속성 값을 `ward`로 설정한 것이다.

덴드로그램에서 최적의 클러스터 수를 찾는 방법은 가장 긴 세로선을 지나는 가로선을 긋는 것이다.

대신 수평 막대에 닿지 않는 선에서다.

애초에 보니까 그냥 덴드로그램에서 색깔로 클러스터를 구분해 주는 것으로 보인다.

위와 같은 경우엔 최적의 클러스터 개수가 두 개가 나올 수 있는데, 3 아니면 5다.

이전 k-means에선 WCSS를 진행한 결과 최적의 군집 수가 5가 나왔으므로, 5가 최적인 것으로 판단하고 실습을 진행한다. 

## Training the Hierarchical Clustering model on the dataset

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters = 5, metric = 'euclidean', linkage = 'ward')
y_hc = hc.fit_predict(X)
```

`metric`: 거리의 종류

클러스터 사이의 분산을 얻기 위해 사용한다.

`linkage`: 덴드로그램 생성했을 때의 `method` 속성과 같은 역할이다.

```python
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], color='red', label='Cluster 1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], color='orange', label='Cluster 2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], color='green', label='Cluster 3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], color='blue', label='Cluster 4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], color='purple', label='Cluster 5')

plt.title('Clusters of customers')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.legend()
plt.show()
```
