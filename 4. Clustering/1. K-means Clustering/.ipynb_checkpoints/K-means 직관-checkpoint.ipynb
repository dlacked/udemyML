# Clustering

## 지도학습과 비지도 학습
지도학습은 회귀와 분류 모델에 해당한다.

이는 학습 데이터에 정답이 이미 존재한다는 것으로, 모델에게 정답이 있는 내용을 학습할 기회를 준다.

하지만 비지도 학습 머신 러닝은 학습 데이터에 정답이 주어지지 않아 모델이 스스로 생각해야 한다.

예를 들어, 레이블이 없는 이미지가 모두 포함된 입력 데이터를 모델에게 제공한 다음

과일들을 서로 다른 카테고리로 나눠 그룹화하라고 명령한다.

일종의 유사성과 차이점이 데이터에 존재한다는 것만 인지한 상태에서

모델은 스스로 결론을 내리고 직접 그룹을 만들어야 한다.

## 군집화

군집화는 비지도 학습 기법으로, 레이블이 지정되지 않은 데이터를 그룹화한다.

## K-Means Clustering

1. 원하는 군집 수를 설정한다.

2. 산점도에 중심점을 무작위로 배치한다.

3. K-means는 등거리 직선을 그려 분류를 시작한다.

4. 각 군집의 모든 X좌표와 Y좌표를 가져와서 평균을 구하면 질량 중심의 위치가 나온다.

5. 그 곳으로 중심점을 옮긴다.

6. 3번에서 5번 과정을 반복한다.

7. 반복하여 변화가 생기지 않으면 분류가 끝난다.

## The Elbow Method

Elbow Method를 통해 군집 최적 개수를 찾을 수 있다.

WCSS 공식을 따르는데, 공식은 다음과 같다.

$$WCSS=\sum{}{P_i in Cluster 1}distance(P_i, C_1)^2 + \sum{}{P_i in Cluster 2}distance(P_i, C_2)^2 + ...$$

군집이 한 개 있다면, 각 점과 해당하는 중심점 사이의 거리를 측정하고 제곱한 다음에 합산하기만 하면 된다.

군집이 두 개 있다면(A, B), A 점끼리 계산하고 B 점끼리 계산한 후 더한다.

이런 과정을 위해선 Elbow Method를 먼저 진행하고 K-means를 구하는 게 아닌,

모든 셋업마다 K-means를 수차례 진행하며 WCSS를 구해야 Elbow Method를 진행할 수 있다.

군집이 많을수록 WCSS의 값이 작아지는데,

WCSS의 값이 급격하게 떨어지는, 즉 절단점이 곧 군집의 최적 개수, Elbow Method가 된다.

## K-means++

동일한 머신러닝 모델로 K-평균을 두 차례 실행할 때 결과는 서로 다르다.

중심점을 초기에 어디에 설정했냐에 따라 계산값도 달라지기 때문이다.

이것을 무작위 초기화의 함정이라고 하며, 바람직하지 않은 부분이다.

이를 방지하기 위해 K-means++가 출현하였다.

기본적으로 K-means와 동일하지만, 특정한 방식으로 중심점 초기화를 시작할 때 일부 단계가 추가된다.

1. 산점도에 중심점 하나를 무작위로 배치한다.

2. 모든 데이터 포인트를 기준으로 선택된 중심점 중 가장 가까운 중심점과의 거리를 제각각 계산한다.

3. 가중치 무작위 선택을 사용해 기존 중심점과의 거리의 제곱을 기준으로 다음 중심점을 선택한다.

4. 2번에서 3번 과정을 반복한다.

5. 반복하여 변화가 생기지 않으면 분류가 끝난다.

K-means++를 쓴다고 해서 초기화로 인한 문제가 아예 발생하지 않는다고 할 수는 없다.

하지만 가중치가 부여된 무작위 방식이다보니 문제가 발생할 가능성이 훨씬 낮다는 게 장점이다.
